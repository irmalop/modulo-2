Joins

```python
bd_veterinario = pd.read_csv('data/bd_veterinario (1).csv')
bd_veterinario
```

```python
bd_propietarios = pd.read_csv('data/bd_propetarios (1).csv')
bd_propietarios
```

Left Join

```python
pd.merge(bd_veterinario, bd_propietarios, how = 'left', on = 'id_propetario')
```

Right Join

```python
pd.merge(bd_veterinario, bd_propietarios, how = 'right', on = 'id_propetario')
```

Inner Join

```python
pd.merge(bd_veterinario, bd_propietarios, how = 'inner', on = 'id_propetario')
```

Outer Join

```python
pd.merge(bd_veterinario, bd_propietarios, how = 'outer', on = 'id_propetario')
```
